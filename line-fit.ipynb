# Line fitting example

```python
import random

import numpy as np
import matplotlib.pyplot as plt

from lin_regression import LinearRegression
```

```python
def f(x):
    return 2 + 1.2 * x
```

```python
n = 7
```

```python
X = np.arange(start=1, stop=n + 1).reshape(-1, 1)
y = f(X) + np.random.normal(loc=0, scale=0.5, size=X.shape)
```

```python
plt.figure(figsize=(8, 6))

plt.plot(X, y, 'bo')
plt.axis(
    xmin=0, xmax=n + 1,
    ymin=f(0) - 1, ymax=f(n + 1) + 1
)

plt.show()
```

```python
linReg = LinearRegression()
```

```python
linReg.fit(X, y)
```

```python
print(linReg.coef_)
```

```python
X_line = np.arange(start=0, stop=n + 2).reshape(-1, 1)
y_line = linReg.predict(X_line)
```

```python
plt.figure(figsize=(8, 6))

plt.plot(X_line, y_line, linewidth=2.0)
plt.plot(X, y, 'bo')
plt.axis(
    xmin=0, xmax=n + 1,
    ymin=f(0) - 1, ymax=f(n + 1) + 1
)

plt.show()
```
